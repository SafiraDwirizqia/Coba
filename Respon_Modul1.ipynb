### Import Library

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

### Operasi Grayscale

```python
#image = plt.imread('image/foto.jpg')
image = plt.imread('D:/SEMESTER 4/Pengolahan Citra Digital/Praktikum/Modul 1/respon-modul-1-RegiaPuspaAmaranthi/image/foto.jpg')
plt.subplot(121)
plt.title("Original image")
plt.imshow(image, cmap="gray")

gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.subplot(122)
plt.title("Gray image")
plt.imshow(gray_image,cmap="gray")
```

### 1.	Lakukan normalisasi terhadap citra dan tunjukkan histogram hasil normalisasi

```python
def normalisasi(gray_image):
    min_value = np.min(gray_image) 
    max_value = np.max(gray_image) 
    range_value = max_value - min_value

    normalized_image = ((gray_image - min_value) / range_value) * 255
    return normalized_image.astype(np.uint8)

gray_image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
normalized_image = normalisasi(gray_image)

gray_image.sum()

kanvas = [i for i in range (256)]

height, width = normalized_image.shape

print("Tinggi: ", height)
print("Lebar: ", width)

kanvas2 = np.zeros(256, dtype=int)

for i in range(height):
    for j in range(width):
        value = gray_image[i,j]
        kanvas2[value] += 1

print(kanvas2)

plt.bar(kanvas, kanvas2, color="blue")
```

```python
height, width = normalized_image.shape
normalized_image.shape

n = height * width
print(n)

normalized_image = [i for i in range(256)]
print(normalized_image)

hi = kanvas2/n
plt.ylabel("Normalized Frequency")
plt.bar(normalized_image, hi, color='blue')
```

### 2.	Translasi dengan x = 100, y = -90

```python
translasi = np.zeros((height, width)).astype(int)
for x in range(height):
    for y in range(width):
        new_x = x + 100
        new_y = y - 90

        if new_x >= 0 and new_x < height and new_y >= 0 and new_y < width:
            translasi[new_x][new_y] = gray_image[x,y]

plt.subplot(121)
plt.title("Gray image")
plt.imshow(gray_image, cmap='gray')

plt.subplot(122)
plt.title("Translasi")
plt.imshow(translasi, cmap='gray')
```

### 3.	Rotasi 110`(berlawanan jarum jam)

```python
sudut_rotasi = np.radians(110)
cosine = np.cos(sudut_rotasi)
sine = np.sin(sudut_rotasi)
height, width = gray_image.shape
rotasi = np.zeros((height, width), dtype = np.uint8)
x_tengah = width / 2
y_tengah = height / 2
#titik_pusat = (height/2, width/2)

for x in range(height):
    for y in range(width):
        xp = int((x - x_tengah) * cosine - (y - y_tengah) * sine + x_tengah)
        yp = int((x - x_tengah) * sine + (y - y_tengah) * cosine + y_tengah)
        if xp >= 0 and y >= 0 and xp < height and yp < width:
        #0 <= xp < height and 0 <= yp < width:
            rotasi[xp, yp]= gray_image[x, y]

plt.subplot(121)
plt.title("Gray image")
plt.imshow(gray_image, cmap='gray')

plt.subplot(122)
plt.title("Rotasi")
plt.imshow(rotasi, cmap='gray')
```

### 4.	Refleksi sumbu x

```python
cermin_x = np.zeros(gray_image.shape)
height, width = gray_image.shape
for i in range(height):
    for j in range(width):
        cermin_x[height-1-i][j] = gray_image[i][j]

plt.title("Pencerminan Sumbu X")
plt.imshow(cermin_x, cmap='gray')
```

### 5.	Dilatasi dengan skala = -3

```python
def dilatasi(gray_image, skala):
    height, width = gray_image.shape

    new_height = int(height / abs(skala))
    new_width = int(width / abs(skala))
    
    dilated_image = np.zeros((new_height, new_width), dtype=np.uint8)
    
    for i in range(new_height):
        for j in range(new_width):
            new_i = int(i * abs(skala))
            new_j = int(j * abs(skala))
            if new_i < height and new_j < width:
                dilated_image[i, j] = gray_image[new_i, new_j]
    
    return dilated_image

skala = -3
dilated_image = dilatasi(gray_image, skala)

plt.title("Dilatasi")
plt.imshow(dilated_image, cmap="gray")
```

### 6.	Ubah citra ke dalam bentuk negatif, lalu bandingkan dengan histogram hasil normalisasi.

```python
#citra = cv.imread('image/foto.jpg')
citra = cv.imread('D:/SEMESTER 4/Pengolahan Citra Digital/Praktikum/Modul 1/respon-modul-1-RegiaPuspaAmaranthi/image/foto.jpg')

ukuran_horizontal = citra.shape[1]
ukuran_vertikal = citra.shape[0]

for x in range(ukuran_horizontal):
    for y in range(ukuran_vertikal):
        R = 255 - citra[y, x][2]  
        G = 255 - citra[y, x][1]  
        B = 255 - citra[y, x][0]  
        citra[y, x] = [B, G, R]

plt.imshow(cv.cvtColor(citra, cv.COLOR_BGR2GRAY))
plt.title('Citra Negatif')
plt.show()
```
